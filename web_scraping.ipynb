```python
from bs4 import BeautifulSoup
```

```python
import os
```

```python
from urllib.request import Request, urlopen
```

```python
url = 'https://www.gettyimages.com/photos/cristiano-ronaldo-?assettype=image&license=rf&alloweduse=availableforalluses&family=creative&phrase=cristiano%20ronaldo%20&sort=mostpopular'

headers = requests.utils.default_headers()
headers.update({'User-Agent':'your Agent'})

r = requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(r.text,'html.parser')
```

```python
soup
```

```python
images = soup.find_all('img')
```

```python
print(len(images))
```

```python
for image in images:
  try:
    print(image['src'])
    print(image['alt'])
  except:
    continue
```

```python
i=0
for image in images:
  try:
    if 'jpg' in image['src']:
      with open(f'image_{i}.jpg','wb') as f:
        im = requests.get(image['src'])
        f.write(im.content)


      i+=1
      # break
  except:
    continue
```
